### Functions For All Methods

#### 1. Bisection Method

```python
def bisectionMethod(a,b,t):
    niter=0
    while (abs(a-b)>=t):
        mid=(a+b)/2.0
        prod=f(a)*f(mid)
        if prod>t:
            a=mid
        else:
            if prod<t:
                b=mid
        niter+=1 
        #print(niter, a, b, mid, f(a), f(b), f(mid), ((abs(a-b)/a)*100))
    return mid,niter

def bisectionMethod_SBS(a,b,t):
    niter=0
    while (abs(a-b)>=t):
        mid=(a+b)/2.0
        prod=f(a)*f(mid)
        if prod>t:
            a=mid
        else:
            if prod<t:
                b=mid
        niter+=1 
        #print(niter, a, b, mid, f(a), f(b), f(mid), ((abs(a-b)/a)*100))
        list_output.insert(END,'Root: ',mid)
        list_output.insert(END,'Iterations: ',niter)
        list_output.insert(END,'--------------------------------------------')
    return mid,niter
```

#### 2. Regula Farsi Method

```python
def regula_farsi(a,b,niter,tol,verbose=False):
    if f(a) * f(b)>0:
        c=None
        msg="Invalid points"
    else:
        for i in range(niter):
            fa=f(a)
            fb=f(b)
            c=(a*fb-b*fa)/(fb-fa)
            fc=f(c)
            if verbose:
                print("Root is {} after {} iterations".format(c,i))
            msg="Maximum iterations completed"
            if abs(fc)<tol:
                msg="Root found!!"
                break
            elif fa*fc<0:
                b=c
            elif fb*fc<0:
                a=c
    return c,i

def regula_farsi_SBS(a,b,niter,tol,verbose=False):
    if f(a) * f(b)>0:
        c=None
        msg="Invalid points"
    else:
        for i in range(niter):
            fa=f(a)
            fb=f(b)
            c=(a*fb-b*fa)/(fb-fa)
            fc=f(c)
            list_output.insert(END,'Root: ',c)
            list_output.insert(END,'Iterations: ',i)
            list_output.insert(END,'--------------------------------------------')
            if verbose:
                print("Root is {} after {} iterations".format(c,i))
            msg="Maximum iterations completed"
            if abs(fc)<tol:
                msg="Root found!!"
                break
            elif fa*fc<0:
                b=c
            elif fb*fc<0:
                a=c
    
    return c,i
```

#### 3. Secant Method

```python
def secant(x0,x1,tol,niter):
    for it in range(niter):
        x2=x0-(x1-x0)/(f(x1)-f(x0))*f(x0)
        if abs(x2-x1)<tol: break
        else:
            x0=x1
            x1=x2
    else:
        print("Maximum iteratons finished and secant method is diverging")
    return x2,it

def secant_SBS(x0,x1,tol,niter):
    for it in range(niter):
        x2=x0-(x1-x0)/(f(x1)-f(x0))*f(x0)
        list_output.insert(END,'Root: ',x2)
        list_output.insert(END,'Iterations: ',it)
        list_output.insert(END,'--------------------------------------------')
        if abs(x2-x1)<tol: break
        else:
            x0=x1
            x1=x2
    else:
        print("Maximum iteratons finished and secant method is diverging")
    return x2,it            
```

#### 4. Muller's Method

```python
def mullersMethod(x,x1,x3,tol):
    i = 0
    while(abs(f(x)) > tol):
        q = (x - x1)/(x1 - x3)
        a = q*f(x) - q*(1+q)*f(x1) + q**2*f(x3)
        b = (2*q + 1)*f(x) - (1+q)**2*f(x1) + q**2*f(x3)
        c = (1 + q)*f(x)
        r = x - (x - x1)*((2*c)/(b + mm.sqrt(b**2 - 4*a*c)))
        s = x - (x - x1)*((2*c)/(b - mm.sqrt(b**2 - 4*a*c)))
        if (abs(f(r)) < abs(f(s))): xNew = r
        else: xNew = s
        x3 = x1
        x1 = x
        x = xNew
        i = i + 1
    return xNew,i

def mullersMethod_SBS(x,x1,x3,tol):
    i = 0
    while(abs(f(x)) > tol):
        q = (x - x1)/(x1 - x3)
        a = q*f(x) - q*(1+q)*f(x1) + q**2*f(x3)
        b = (2*q + 1)*f(x) - (1+q)**2*f(x1) + q**2*f(x3)
        c = (1 + q)*f(x)
        r = x - (x - x1)*((2*c)/(b + mm.sqrt(b**2 - 4*a*c)))
        s = x - (x - x1)*((2*c)/(b - mm.sqrt(b**2 - 4*a*c)))
        if (abs(f(r)) < abs(f(s))): xNew = r
        else: xNew = s
        x3 = x1
        x1 = x
        x = xNew
        i = i + 1
        list_output.insert(END,'Root: ',xNew)
        list_output.insert(END,'Iterations: ',i)
        list_output.insert(END,'--------------------------------------------')
    return xNew,i
```

#### 5. Newton Raphson Method

```python
def newton_raphson(x0,tol,niter):
    for it in range(niter):
        xNew = x0-(f(x0)/der_f(x0))
        #print(it,'----',xNew)
        if abs(xNew-x0)<tol: 
            break
        else:
            x0 = xNew
    else:
        print("Maximum iteratons finished and Newton Raphson method is diverging")
    return xNew,it         

def newton_raphson_SBS(x0,tol,niter):
    for it in range(niter):
        xNew = x0-(f(x0)/der_f(x0))
        #print(it,'----',xNew)
        list_output.insert(END,'Root: ',xNew)
        list_output.insert(END,'Iterations: ',it)
        list_output.insert(END,'--------------------------------------------')
        if abs(xNew-x0)<tol: 
            break
        else:
            x0 = xNew
    else:
        print("Maximum iteratons finished and Newton Raphson method is diverging")
    return xNew,it         
```
